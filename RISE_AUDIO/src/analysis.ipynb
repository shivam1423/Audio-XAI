```python
#now based on the codebase read single audio file,

import argparse
import numpy as np
import torch
import matplotlib.pyplot as plt

from src.audio_io import AudioProcessor
from src.feature_extractor import process_masked_audio_list
from src.utils import MEL_SR, MEL_N_FFT, MEL_HOP_LENGTH, MEL_N_MELS


def spec_to_db_for_plot(spec_complex: torch.Tensor) -> np.ndarray:
    """
    Convert complex spectrogram (1, F, T) to dB (for visualization only).
    Uses magnitude, normalized to [min, max] dB for plotting.
    """
    # magnitude
    mag = torch.abs(spec_complex).squeeze(0)  # (F, T)
    mag_np = mag.detach().cpu().numpy().astype(np.float32)

    # convert to dB-like scale safely
    eps = 1e-10
    db = 20.0 * np.log10(np.maximum(mag_np, eps))
    # normalize dB for visualization
    db_norm = (db - db.min()) / (db.max() - db.min() + 1e-12)
    return db_norm
```

```python
def main():
    
    # 1) Read + resample using AudioProcessor
    ap = AudioProcessor()
    audio, sr = ap.load_audio("test_audio/4-195497-A-11.wav")  # resampled to TARGET_ORIGINAL_SR
    print(audio.shape,sr)
    n_samples = audio.shape[1]

    # 2) Convert to complex spectrogram
    spec = ap.audio_to_spectrogram(audio)  # (1, F, T)
    print(spec.shape)
    # 3) Prepare a single mask and apply it
    #    Ensure masks exist for current spectrogram size and fetch one mask
    ap.generate_or_load_masks(spec.shape[1:], n_masks=6000)  # create/load file; we only need one
    mask = ap.get_mask(3050)  # shape (F, T)
    real_imag = ap.complex_to_real_imag(spec)  # (2, F, T)
    masked_real_imag = ap.apply_masking(real_imag, mask)  # (2, F, T)
    masked_spec = ap.real_imag_to_complex(masked_real_imag)  # (1, F, T)
    print(real_imag.shape)
    # 4) Reconstruct masked audio (so we can compute log-mel on it)
    masked_audio = ap.spectrogram_to_audio(masked_spec, n_samples)  # (1, n_samples)
    
    # 5) Compute log-mel spectrogram image using existing function
    mel_imgs = process_masked_audio_list(
        [masked_audio],
        sr=MEL_SR,
        n_fft=MEL_N_FFT,
        hop_length=MEL_HOP_LENGTH,
        n_mels=MEL_N_MELS,
        orig_sr=sr
    )
    mel_img = mel_imgs[0]  # PIL.Image (grayscale, 224x224)

    # 6) Visualize: resampled spectrogram, masked spectrogram, and log-mel spectrogram
    resampled_db = spec_to_db_for_plot(spec)           # (F, T)
    masked_db = spec_to_db_for_plot(masked_spec)       # (F, T)
    mel_np = np.array(mel_img)                         # (224, 224), uint8

    fig, axes = plt.subplots(1, 3, figsize=(15, 4), constrained_layout=True)

    im0 = axes[0].imshow(resampled_db, origin="lower", aspect="auto", cmap="gray")
    axes[0].set_title("Resampled Spectrogram 48Khz")
    axes[0].set_xlabel("Time")
    axes[0].set_ylabel("Frequency")
    fig.colorbar(im0, ax=axes[0], fraction=0.046, pad=0.04)

    im1 = axes[1].imshow(masked_db, origin="lower", aspect="auto", cmap="gray")
    axes[1].set_title("Masked Spectrogram")
    axes[1].set_xlabel("Time")
    axes[1].set_ylabel("Frequency")
    fig.colorbar(im1, ax=axes[1], fraction=0.046, pad=0.04)

    im2 = axes[2].imshow(mel_np, origin="lower", aspect="auto", cmap="gray")
    axes[2].set_title("Log-Mel Spectrogram")
    axes[2].set_xlabel("Time")
    axes[2].set_ylabel("Mel bins")
    fig.colorbar(im2, ax=axes[2], fraction=0.046, pad=0.04)

    plt.show()


if __name__ == "__main__":
    main()
```

```python
import numpy as np

def coverage_stats(masks):  # masks: (N,1,H,W) or (N,H,W)
    if masks.ndim == 4: masks = masks[:,0]
    H, W = masks.shape[1], masks.shape[2]
    hits = masks.sum(axis=0)  # (H,W)
    covered = (hits > 0)
    return {
        "coverage_ratio": covered.mean(),
        "mean_hits": hits.mean(),
        "hits_std": hits.std(),
        "row_coverage_mean": (covered.mean(axis=1)).mean(),
        "col_coverage_mean": (covered.mean(axis=0)).mean(),
    }

# Example:
# gen = TFMaskGenerator((1537, 157), soft_masking="none")
# masks = gen.generate_tf_masks(N=6000, single_block=True)  # (6000,1,1537,157)
masks = np.load('results/masks/masks_6000_1537_157_none_single_all.npy')
print(coverage_stats(masks))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Read the mask file
mask_file = '/Users/shivampandey/SS 25/Thesis/RISE_dev/RISE_audio/results/masks/masks_8_none_all.npy'
masks = np.load(mask_file)

print(f"Mask shape: {masks.shape}")
print(f"Mask data type: {masks.dtype}")
print(f"Min value: {masks.min()}")
print(f"Max value: {masks.max()}")

# Remove the singleton dimension (8, 1, 1537, 144) -> (8, 1537, 144)
masks_squeezed = np.squeeze(masks, axis=1)
print(f"Squeezed shape: {masks_squeezed.shape}")

# Plot the first mask
plt.figure(figsize=(12, 8))
plt.imshow(masks_squeezed[0], cmap='viridis', aspect='auto', origin='lower')
plt.colorbar(label='Mask Value')
plt.title('First Mask (Frequency vs Time)')
plt.xlabel('Time Frames')
plt.ylabel('Frequency Bins')
plt.show()

# Plot all 8 masks in a grid
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()

for i in range(8):
    im = axes[i].imshow(masks_squeezed[i], cmap='viridis', aspect='auto', origin='lower')
    axes[i].set_title(f'Mask {i}')
    axes[i].set_xlabel('Time Frames')
    axes[i].set_ylabel('Frequency Bins')
    plt.colorbar(im, ax=axes[i])

plt.tight_layout()
plt.show()

# Plot the average of all masks
plt.figure(figsize=(12, 8))
avg_mask = np.mean(masks_squeezed, axis=0)
plt.imshow(avg_mask, cmap='viridis', aspect='auto', origin='lower')
plt.colorbar(label='Average Mask Value')
plt.title('Average of All Masks')
plt.xlabel('Time Frames')
plt.ylabel('Frequency Bins')
plt.show()
```
